Importa las librerías necesarias (pandas, numpy, matplotlib.pyplot).

```python
#LIBRERIAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Imprime las últimas 5 filas del dataset y genera un resumen estadístico descriptivo de las columnas numéricas.

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df =pd.read_csv(url)
df.head()
df.info()
```

Eliminar datos nulos del DataFrame

```python
df = df[(df[["x","y","z"]]!= 0).all(axis=1)]
```

Crea una nueva columna llamada volumen_aprox calculando el volumen espacial de cada pieza (multiplicando las columnas x * y * z).


```python
df["volumen_aprox"] = df["x"] * df["y"] * df["z"]
df.info()
```

Gráfico de caja (Boxplot)

```python
plt.boxplot(df["price"])
plt.title("Precio", color= "green")
plt.ylabel("Precio", color= "blue")
plt.show()
```

Gráfico de dispersión (Scatter)

```python
plt.scatter(df["volumen_aprox"], df["price"])
plt.title("Volumen Aproximado vs Precio", color= "purple")
plt.xlabel("Volumen Aproximado", color= "green")
plt.ylabel("Precio", color= "yellow")
plt.show
```

Gráfico de Barras

```python
conteo_corte = df["cut"].value_counts()

plt.bar(conteo_corte.index, conteo_corte.values)
plt.title("Cantidad de piezas para cada corte", color= "black")
plt.xlabel("Corte", color= "black")
plt.ylabel("Cantidad", color= "black")
plt.show()
```
